```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
ax = sns.distplot(tmdb.vote_average)
ax.set(xlabel='Nota média', ylabel='Densidade', title='Média de votos em filmes no TMDB 5000')
```

```python
ax = sns.distplot(tmdb.vote_average, norm_hist=False, kde=False)
ax.set(xlabel='Nota média', ylabel='Frequencia', title='Média de votos em filmes no TMDB 5000')
```

```python
ax = sns.boxplot(x=tmdb.vote_average)
ax.set(xlabel='Nota média dos filmes')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000')
```

### Detectamos que alguns filmes tiveram poucos votos, ou nenhum voto. Decidimos que esses filmes não devem pertencer a nossa análise 

```python
tmdb_com_mais_de_10_votos = tmdb.query('vote_count >= 10')
tmdb_com_mais_de_10_votos.describe()
```

```python
ax = sns.distplot(tmdb_com_mais_de_10_votos.vote_average, norm_hist=False, kde=False)
ax.set(xlabel='Nota média', ylabel='Frequencia', title='Média de votos em filmes no TMDB 5000 dentre os filmes com 10 ou mais votos')
```

```python
ax = sns.distplot(tmdb_com_mais_de_10_votos.vote_average)
ax.set(xlabel='Nota média', ylabel='Densidade', title='Média de votos em filmes no TMDB 5000 dentre os filmes com 10 ou mais votos')
```

```python
ax = sns.boxplot(x=tmdb_com_mais_de_10_votos.vote_average)
ax.set(xlabel='Nota média dos filmes')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000 com 10 ou mais votos')
```

## Analisaremos também o movielens

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean()['rating']
nota_media_por_filme.head()
```

```python
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel='Nota média', ylabel='Densidade', title='Média de votos em filmes no Movielens 100k')
```

```python
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query('rating >= 10').index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel='Nota média', 
       ylabel='Densidade', 
       title='Média de votos em filmes no Movielens 100k com 10 ou mais votos')
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values,
                  hist_kws={'cumulative': True},
                  kde_kws={'cumulative': True},)
ax.set(xlabel='Nota média', 
       ylabel='% de filmes (acumulada)', 
       title='Média de votos em filmes no Movielens 100k com 10 ou mais votos')
ax.grid(True)
```

```python
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel='Nota média dos filmes')
ax.set_title('Distribuição de nota média no Movielens 100k com 10 ou mais votos')
```
